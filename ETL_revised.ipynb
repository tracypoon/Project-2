```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import happy_cities.csv
csv_file = "Data/happy_cities_df.csv"
happy_cities_df = pd.read_csv(csv_file)
happy_cities_df.head()
```

```python
# Import city_weather.csv
csv_file = "new_city_weather.csv"
city_weather = pd.read_csv(csv_file)
city_weather.head()
```

```python
# Drop the first column in happy_cities_df
df1=happy_cities_df.drop(columns=["Unnamed: 0"])
df1
```

```python
# Rename the column names with _ and lower case
df1=df1.rename(columns={"Overall Rank":"overall_rank", "City":"city", "Total Score":"total_score", 
"Emotional & Physical Well-Being": "emotional_physical_wellbeing", "Income & Employment": "income_employment" ,
"Community & Environment": "community_environment", "State":"state"})
df1
```

```python
# Bring the "state" column next to "city"
df1=df1[["overall_rank", "city", "state", "total_score", "emotional_physical_wellbeing", "income_employment" ,
"community_environment"]]
df1
```

```python
# Drop the first column in city_weather
df2=city_weather.drop(columns=["Unnamed: 0"])
df2
```

```python
# Rename columns to all lower case
df2=df2.rename(columns={"City":"city", "State":"state", "Max_temp": "max_temp", "Humidity": "humidity", "Cloudiness": "cloudiness" ,
"Windspeed": "windspeed", "Date":"date", "Time":"time"})
df2
```

```python
# Connect with local database
rds_connection_string = "postgres:postgres@localhost:5432/happycity_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Load df1 into db
df1.to_sql(name='happy_cities', con=engine, if_exists='append', index=False)
```

```python
# Load df2 into db
df2.to_sql(name='city_weather', con=engine, if_exists='append', index=False)
```
