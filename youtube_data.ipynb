```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import first csv file
csv_file = "US_Videos.csv"
us_videos_df = pd.read_csv(csv_file)
us_videos_df.head()
```

```python
# Import second csv file
csv_file = "Youtube_Dataset.csv"
youtube_df = pd.read_csv(csv_file)
youtube_df.head()
```

```python
# Clean up the identify the columns we want from first csv
us_videos_df = us_videos_df[["title", "likes"]].copy()
us_videos_df.head()
```

```python
# Clean up the identify the columns we want from second csv
youtube_df = youtube_df[["Video Name", "Likes"]].copy()
youtube_df.head()
```

```python
# Change the cloumn name of Video Name
youtube_df = youtube_df.rename(columns={"Video Name": "video_name", "Likes": "likes"})
youtube_df
```

```python
# Connect to local database
rds_connection_string = "postgres:postgres@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Use pandas to load first csv converted DataFrame into database
us_videos_df.to_sql(name='videos', con=engine, if_exists='append', index=False)
```

```python
# Use pandas to load second csv converted first DataFrame into database
youtube_df.to_sql(name='youtube', con=engine, if_exists='append', index=False)
```
